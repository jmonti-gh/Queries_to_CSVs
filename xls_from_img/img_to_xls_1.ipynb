1. Instalación de librerías:

Primero, necesitas instalar las librerías necesarias:

Bash

pip install pytesseract Pillow
También necesitas instalar Tesseract OCR, que es el motor de reconocimiento óptico de caracteres que pytesseract utiliza. Puedes descargarlo desde https://github.com/UB-Mannheim/tesseract/wiki (elige la versión adecuada para tu sistema operativo).

2. Código Python:

Python

import pytesseract
from PIL import Image
import csv
import io

# Configura la ruta a tu ejecutable de Tesseract OCR (reemplaza con tu ruta)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Ejemplo en Windows

def extraer_texto_de_imagen(ruta_imagen):
    """Extrae el texto de una imagen usando Tesseract OCR."""
    try:
        imagen = Image.open(ruta_imagen)
        texto = pytesseract.image_to_string(imagen, lang='spa')  # 'spa' para español
        return texto
    except Exception as e:
        print(f"Error al procesar la imagen: {e}")
        return None

def texto_a_csv(texto, nombre_archivo_csv):
    """Convierte el texto extraído en un archivo CSV."""
    if not texto:
        return

    filas = []
    for linea in io.StringIO(texto):
        linea = linea.strip()
        if linea:
            filas.append(linea.split('\t'))  # Asume que las columnas están separadas por tabulaciones

    with open(nombre_archivo_csv, mode='w', newline='', encoding='utf-8') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerows(filas)

    print(f"Archivo CSV '{nombre_archivo_csv}' creado exitosamente.")

# Ruta de la imagen que quieres convertir
ruta_imagen = 'ruta/a/tu/imagen.png'  # Reemplaza con la ruta de tu imagen

# Extrae el texto de la imagen
texto_extraido = extraer_texto_de_imagen(ruta_imagen)

# Convierte el texto a CSV
if texto_extraido:
    texto_a_csv(texto_extraido, 'tabla_extraida.csv')
Explicación:

Importa librerías: Importa pytesseract, Pillow (para abrir la imagen), csv (para trabajar con archivos CSV) e io (para tratar el texto como un archivo).
Configura la ruta de Tesseract: Debes configurar la variable pytesseract.pytesseract.tesseract_cmd con la ruta al ejecutable de Tesseract en tu sistema.
Función extraer_texto_de_imagen:
Abre la imagen usando PIL.Image.open().
Utiliza pytesseract.image_to_string() para extraer el texto de la imagen. El argumento lang='spa' especifica que el texto está en español.
Maneja posibles errores al procesar la imagen.
Función texto_a_csv:
Divide el texto extraído en líneas y luego cada línea en columnas, asumiendo que las columnas están separadas por tabulaciones (\t). Si el separador es diferente, ajústalo en linea.split().
Escribe las filas en un archivo CSV usando el módulo csv.
Ejecución:
Reemplaza 'ruta/a/tu/imagen.png' con la ruta real de tu imagen.
Llama a las funciones extraer_texto_de_imagen y texto_a_csv para realizar la conversión.
Nota:

La precisión de la extracción de texto depende de la calidad de la imagen y la claridad del texto.
Ajusta el argumento lang en pytesseract.image_to_string() si el texto está en otro idioma.
Si las columnas no están separadas por tabulaciones, ajusta el separador en linea.split().
Si tienes problemas con la instalación de Tesseract o pytesseract, consulta la documentación de las librerías para obtener ayuda.
