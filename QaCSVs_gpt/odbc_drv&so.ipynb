### ‚úÖ Programa Python: Mostrar tabla tipo "Drivers" de ODBC

```python
import winreg
import os
import platform
import datetime
from tabulate import tabulate

def get_driver_info():
    drivers = []

    registry_paths = [
        (winreg.HKEY_LOCAL_MACHINE, r"SOFTWARE\ODBC\ODBCINST.INI"),                # 64-bit
        (winreg.HKEY_LOCAL_MACHINE, r"SOFTWARE\WOW6432Node\ODBC\ODBCINST.INI")     # 32-bit
    ]

    for hive, base_path in registry_paths:
        try:
            base_key = winreg.OpenKey(hive, base_path)
            i = 0
            while True:
                try:
                    subkey_name = winreg.EnumKey(base_key, i)
                    if subkey_name != "ODBC Drivers":
                        subkey_path = base_path + "\\" + subkey_name
                        subkey = winreg.OpenKey(hive, subkey_path)

                        # Extraer informaci√≥n b√°sica
                        driver_name = subkey_name
                        driver_dll = _get_registry_value(subkey, "Driver")
                        driver_version = _get_registry_value(subkey, "Version") or ""
                        driver_company = _get_registry_value(subkey, "CompanyName") or ""
                        
                        # Obtener fecha del archivo DLL si existe
                        driver_date = ""
                        if driver_dll and os.path.isfile(driver_dll):
                            timestamp = os.path.getmtime(driver_dll)
                            driver_date = datetime.datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')

                        drivers.append([driver_name, driver_version, driver_company, driver_dll, driver_date])
                        i += 1
                except OSError:
                    break
        except FileNotFoundError:
            continue

    return drivers

def _get_registry_value(key, value_name):
    try:
        value, _ = winreg.QueryValueEx(key, value_name)
        return value
    except FileNotFoundError:
        return None

def main():
    print(f"Sistema: {platform.system()} ({platform.architecture()[0]})\n")
    driver_info = get_driver_info()
    headers = ["Name", "Version", "Company", "File", "Date"]
    print(tabulate(driver_info, headers=headers, tablefmt="grid"))

if __name__ == "__main__":
    main()
```

import os
import platform
import datetime
import csv
from tabulate import tabulate
import ctypes
from ctypes import wintypes

MAX_SUBKEYS = 200

def get_file_version_info(file_path):
    try:
        size = ctypes.windll.version.GetFileVersionInfoSizeW(file_path, None)
        if size == 0:
            return "", ""

        res = ctypes.create_string_buffer(size)
        ctypes.windll.version.GetFileVersionInfoW(file_path, 0, size, res)
        r = ctypes.c_void_p()
        l = wintypes.UINT()

        # Language + codepage
        ctypes.windll.version.VerQueryValueW(res, r"\\VarFileInfo\\Translation", ctypes.byref(r), ctypes.byref(l))
        lang, codepage = ctypes.cast(r.value, ctypes.POINTER(wintypes.WORD * 2)).contents

        # Query string values
        def query_value(name):
            sub_block = f"\\StringFileInfo\\{lang:04x}{codepage:04x}\\{name}"
            r_val = ctypes.c_void_p()
            l_val = wintypes.UINT()
            if ctypes.windll.version.VerQueryValueW(res, sub_block, ctypes.byref(r_val), ctypes.byref(l_val)):
                return ctypes.wstring_at(r_val.value)
            return ""

        version = query_value("FileVersion")
        company = query_value("CompanyName")
        return version, company
    except Exception:
        return "", ""

### ‚úÖ ¬øQu√© cambia?

* Ya no dependemos del contenido del registro para los campos `Version` y `CompanyName`.
* Los extraemos directamente del archivo `.dll` del controlador, si existe.
* Eso es mucho m√°s confiable y suele reflejar lo que ves en la pesta√±a "Details" del archivo en el explorador de Windows.

¬øQuer√©s que esta informaci√≥n tambi√©n se exporte a Excel (`.xlsx`)?


````python
import winreg
import os
import platform
import datetime
import csv
from tabulate import tabulate
import ctypes
from ctypes import wintypes

MAX_SUBKEYS = 200

def get_file_version_info(file_path):
    try:
        size = ctypes.windll.version.GetFileVersionInfoSizeW(file_path, None)
        if size == 0:
            return "", ""

        res = ctypes.create_string_buffer(size)
        ctypes.windll.version.GetFileVersionInfoW(file_path, 0, size, res)
        r = ctypes.c_void_p()
        l = wintypes.UINT()

        # Language + codepage
        ctypes.windll.version.VerQueryValueW(res, r"\\VarFileInfo\\Translation", ctypes.byref(r), ctypes.byref(l))
        lang, codepage = ctypes.cast(r.value, ctypes.POINTER(wintypes.WORD * 2)).contents

        # Query string values
        def query_value(name):
            sub_block = f"\\StringFileInfo\\{lang:04x}{codepage:04x}\\{name}"
            r_val = ctypes.c_void_p()
            l_val = wintypes.UINT()
            if ctypes.windll.version.VerQueryValueW(res, sub_block, ctypes.byref(r_val), ctypes.byref(l_val)):
                return ctypes.wstring_at(r_val.value)
            return ""

        version = query_value("FileVersion")
        company = query_value("CompanyName")
        return version, company
    except Exception:
        return "", ""

def get_driver_info():
    drivers = []

    registry_paths = [
        (winreg.HKEY_LOCAL_MACHINE, r"SOFTWARE\ODBC\ODBCINST.INI", "64-bit"),
        (winreg.HKEY_LOCAL_MACHINE, r"SOFTWARE\WOW6432Node\ODBC\ODBCINST.INI", "32-bit")
    ]

    for hive, base_path, arch in registry_paths:
        try:
            base_key = winreg.OpenKey(hive, base_path)
            for i in range(MAX_SUBKEYS):
                try:
                    subkey_name = winreg.EnumKey(base_key, i)
                    if subkey_name != "ODBC Drivers":
                        subkey_path = base_path + "\\" + subkey_name
                        try:
                            subkey = winreg.OpenKey(hive, subkey_path)

                            driver_name = subkey_name
                            driver_dll = _get_registry_value(subkey, "Driver")

                            # Obtener versi√≥n y compa√±√≠a desde el archivo DLL
                            driver_version, driver_company = "", ""
                            driver_date = ""

                            if driver_dll and os.path.isfile(driver_dll):
                                driver_version, driver_company = get_file_version_info(driver_dll)
                                timestamp = os.path.getmtime(driver_dll)
                                driver_date = datetime.datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')

                            drivers.append([
                                driver_name, driver_version, driver_company, driver_dll, driver_date, arch
                            ])
                        except Exception as e:
                            print(f"‚ö†Ô∏è Error al abrir subclave {subkey_name}: {e}")
                except OSError:
                    break
        except FileNotFoundError:
            continue

    return drivers

def _get_registry_value(key, value_name):
    try:
        value, _ = winreg.QueryValueEx(key, value_name)
        return value
    except Exception:
        return None

def export_to_csv(data, headers, filename="odbc_drivers.csv"):
    try:
        with open(filename, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(data)
        print(f"\nüìÑ Archivo CSV exportado como '{filename}'")
    except Exception as e:
        print(f"‚ùå Error al exportar CSV: {e}")

def main():
    print(f"Sistema: {platform.system()} ({platform.architecture()[0]})\n")
    driver_info = get_driver_info()
    headers = ["Name", "Version", "Company", "File", "Date", "Architecture"]
    print(tabulate(driver_info, headers=headers, tablefmt="grid"))

    if driver_info:
        export_to_csv(driver_info, headers)
    else:
        print("\n‚ö†Ô∏è No se encontraron controladores ODBC.")

if __name__ == "__main__":
    main()
````
