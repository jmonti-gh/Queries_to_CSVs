# Código Python para ejecutar una query SQL Server y guardar resultados en CSV

Aquí tienes un código Python que conecta a una base de datos SQL Server, ejecuta una consulta SQL y guarda los resultados en un archivo CSV con las especificaciones requeridas (separador ";", codificación UTF-8 con BOM):

```python
import pyodbc
import csv

def execute_query_and_save_to_csv(server, database, username, password, sql_query, output_file):
    """
    Ejecuta una consulta SQL en SQL Server y guarda los resultados en un archivo CSV.
    
    Parámetros:
    - server: Nombre del servidor SQL Server
    - database: Nombre de la base de datos
    - username: Nombre de usuario para la autenticación
    - password: Contraseña para la autenticación
    - sql_query: Consulta SQL a ejecutar
    - output_file: Ruta del archivo CSV de salida
    """
    
    # Cadena de conexión
    connection_string = f"DRIVER={{SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}"
    
    try:
        # Establecer conexión
        with pyodbc.connect(connection_string) as conn:
            cursor = conn.cursor()
            
            # Ejecutar la consulta
            cursor.execute(sql_query)
            
            # Obtener los nombres de las columnas
            columns = [column[0] for column in cursor.description]
            
            # Obtener todos los resultados
            rows = cursor.fetchall()
            
            # Escribir los resultados en un archivo CSV
            with open(output_file, 'w', newline='', encoding='utf-8-sig') as csvfile:
                writer = csv.writer(csvfile, delimiter=';')
                
                # Escribir el encabezado (nombres de columnas)
                writer.writerow(columns)
                
                # Escribir las filas de datos
                writer.writerows(rows)
                
        print(f"Los resultados se han guardado correctamente en {output_file}")
        
    except pyodbc.Error as e:
        print(f"Error al conectarse a SQL Server: {e}")
    except Exception as e:
        print(f"Error inesperado: {e}")

# Ejemplo de uso
if __name__ == "__main__":
    # Configuración de conexión (modificar con tus credenciales)
    server = "nombre_servidor"
    database = "nombre_base_datos"
    username = "tu_usuario"
    password = "tu_contraseña"
    
    # Consulta SQL (modificar según necesidades)
    sql_query = "SELECT * FROM TuTabla WHERE condicion = 1"
    
    # Archivo de salida
    output_file = "resultados.csv"
    
    # Ejecutar la función
    execute_query_and_save_to_csv(server, database, username, password, sql_query, output_file)
```

## Características del código

- Conexión segura a SQL Server usando autenticación por usuario/contraseña
- Ejecución de consultas SQL parametrizables
- Generación de CSV con:
  - Delimitador ";" (punto y coma)
  - Codificación UTF-8 con BOM (usando 'utf-8-sig')
  - Nombres de columnas como encabezado
- Manejo básico de errores
