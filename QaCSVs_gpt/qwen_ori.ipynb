Podrías hacer un código python que lance un query SQL a una base de datos SQL Server y que el resultado de dicha query la guarde como un archivo csv separado por ';' codificado utf-8 con BOM

Aquí tienes un script en Python que cumple con tus requisitos. Conecta a una base de datos SQL Server, ejecuta una consulta y guarda los resultados en un archivo CSV con separador `;`, codificación **UTF-8 con BOM**:

```python
import pyodbc
import csv

# Configuración de la conexión a SQL Server
server = 'tu_servidor_sql'  # Ejemplo: 'localhost' o '192.168.1.100'
database = 'tu_base_de_datos'
username = 'tu_usuario'
password = 'tu_contraseña'
port = '1433'  # Puerto predeterminado para SQL Server

# Consulta SQL que deseas ejecutar
query = "SELECT * FROM tu_tabla"

# Ruta y nombre del archivo CSV de salida
archivo_csv = 'salida.csv'

try:
    # Cadena de conexión
    conn_str = (
        f'DRIVER={{ODBC Driver 17 for SQL Server}};'
        f'SERVER={server},{port};'
        f'DATABASE={database};'
        f'UID={username};'
        f'PWD={password}'
    )
    
    # Establecer conexión
    conn = pyodbc.connect(conn_str)
    cursor = conn.cursor()
    
    # Ejecutar consulta
    cursor.execute(query)
    
    # Obtener encabezados y filas
    columnas = [column[0] for column in cursor.description]
    filas = cursor.fetchall()
    
    # Guardar en CSV con codificación UTF-8 con BOM
    with open(archivo_csv, mode='w', newline='', encoding='utf-8-sig') as file:
        writer = csv.writer(file, delimiter=';')
        writer.writerow(columnas)  # Escribir encabezados
        writer.writerows(filas)    # Escribir filas
        
    print(f"Archivo '{archivo_csv}' guardado exitosamente.")

except Exception as e:
    print(f"Error: {str(e)}")

finally:
    # Cerrar conexión
    if 'conn' in locals():
        cursor.close()
        conn.close()
```

---

### **Instrucciones**
1. **Instala `pyodbc`** (si no lo tienes):
   ```bash
   pip install pyodbc
   ```

2. **Reemplaza los siguientes valores en el código**:
   - `tu_servidor_sql`: Dirección IP o nombre del servidor SQL Server.
   - `tu_base_de_datos`: Nombre de la base de datos.
   - `tu_usuario`: Usuario de SQL Server.
   - `tu_contraseña`: Contraseña del usuario.
   - `tu_tabla`: Nombre de la tabla o consulta personalizada.

3. **Ejecuta el script**:
   ```bash
   python script_sql_to_csv.py
   ```

---

### **Explicación clave**
- **Codificación UTF-8 con BOM**: Se usa `encoding='utf-8-sig'` para incluir el BOM (útil para compatibilidad con Excel).
- **Separador `;`**: Se configura con `delimiter=';'` en `csv.writer`.
- **Manejo de errores**: El bloque `try-except` captura errores de conexión o consulta.
- **Cierre seguro**: Se asegura de cerrar la conexión incluso si ocurre un error.

---

