# Try unique in multiples columns

```python
import pandas as pd
```

```python
### Read input xlsx file
fn = 'listas/CSVs_list_min3.xlsx'
df = pd.read_excel(fn)
df
```

```python
### Sep in mini DF by unique Server Base col
#df[['Server', 'Base']].unique()
# AttributeError: 'DataFrame' object has no attribute 'unique'
#https://www.statology.org/pandas-unique-multiple-columns/

print(df[['Server', 'Base']].values, '\n')
print(df[['Server', 'Base']].values.ravel(), '\n')
pd.unique(df[['Server', 'Base']].values.ravel())        # it WORKS!
```

```python
p_sb = pd.unique(df[['Server', 'Base']].values.ravel())
print(len(p_sb))
p_sb
```

```python
### And NOW divde the mini_dfs
mini_dfs = []
tot = len(p_sb)     # must be a pair number of course (ahora no lo cheaqueamos)
for i in range(0, tot, 2):
    print(i, end='  ')
    srv = p_sb[i]
    db = p_sb[i+1]
    mini_df = df[(df.Server == srv) | (df.Base == db)]
    mini_df.reset_index(drop=True, inplace=True)
    mini_dfs.append(mini_df)

for i in range(len(mini_dfs)):
    display(mini_dfs[i])

#// Es mejor usar una sola col Server Base porque sino tengo que completar dos cols y
# el usuario puede haberse olvidado la base y no necesariamente coincide con la arriba
# igual vamos a terminar esto per-joder.
```
